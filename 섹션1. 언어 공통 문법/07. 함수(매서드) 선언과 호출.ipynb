## 07. 함수(매서드) 선언과 호출

### 언어별 공통 문법: 함수(매서드)

### 프로그래밍에서의 함수

**[함수의 정의는 무엇일까?]**
- `함수`는 프로그래밍에서 `특정 작업을 수행하기 위해 독립적으로 작성된 코드 블록`입니다.

- 함수는 코드의 `재사용성`을 높이고, 코드의 `가독성`을 향상시키며, 유지보수를 쉽게 만듭니다.0.

- 함수는 `입력값`을 받아 작업을 수행하고, `결과값`을 반환할 수 있습니다.

**[함수의 예시는 무엇일까?]**
1. 뭔가 더하고 싶다. 매번 더하는 것을 출력문에 작성해서 복사 + 붙여넣기?

2. 뭔가 반복하거나 입력값에 따라 조건을 달리하고 싶다: 매번 반복문과 조건문 작성?

3. 언어를 만든 사람이 미리 정해놓은 기능을 쓰고 싶다: 어떻게 사용하지?

**함수는 3가지만 체크하자**
1. 정의와 호출
    - 함수에게 이름을 지어주고, 기능을 불어넣어준다.

2. 입력값(파라미터)
    - 함수에게 던져주는 값, 입력값이 없는 함수도 만들 수 있다.

3. 리턴값(돌려주는 값, 반환값)
    - 함수가 기능을 모두 마치고 돌려주는 값.
    - 언어에 따라서 돌려주는 값의 자료형을 명시하는 경우가 있다.


```c
#include <stdio.h>

int add(int a, int b){
    return a + b;
}

int main(){
    int result = add(3,5);
    printf("출력값: %d\n", result);
    return 0;
}
```

1. 정의와 호출
    - add라고 이름을 붙여줬다. 그리고 add(3, 5)로 호출했다.

2. 입력값(파라미터)
    - 함수 내에서 a와 b로 쓰이겠다고 정의했다. 이러면 2개의 입력값이 필요하다.

3. 리턴값(돌려주는 값, 반환값)
    - 함수에서는 a + b를 돌려주고 있다.
    - C언어에서 리턴값의 자료형은 함수명 앞에다가 적는다.
    - 여기서는 int로 되어있으니 숫자형을 돌려준다는 뜻이다.

### 함수와 메서드 요약

**[C언어 함수]**
1. 함수 앞에 리턴값 자료형 있다.
2. main이라고 적힌 함수부터 코드가 시작

**[Java 메서드]**
1. 메서드 앞에 뭔가 용어가 많았다.
2. C언어와 같이 리턴값에 대한 자료형이 있다.
3. C언어와 같이 main 함수부터 코드가 시작

**[Python 함수]**
1. 함수를 만들 때 def라는 키워드를 쓴다.
2. 자료형에 대한 힌트가 없다.
3. 함수 내용을 괄호가 아닌 `들여쓰기`로 구분
